```python
import pandas as pd 
import numpy as np 
```

```python
# loading the data 

df = pd.read_csv("transactions.csv")
```

```python
df.head(10)
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# fixing the column names 

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
# converting  date column
df['transactiondate'] = pd.to_datetime(df['transactiondate'], errors='coerce')
```

```python
df['transactiondate'].head(10)
```

```python
df['transactiondate'] = pd.to_datetime(
    df['transactiondate'],
    errors='coerce',
    dayfirst=True
)
```

```python
df['transactiondate'].isnull().sum()
```

```python
df['transactiondate'].head(10)
```

```python
def fix_date(x):
    try:
        return pd.to_datetime(x, dayfirst=True)
    except:
        try:
            return pd.to_datetime(x, dayfirst=False)
        except:
            return pd.NaT

df['transactiondate'] = df['transactiondate'].apply(fix_date)
```

```python
df['transactiondate'].head(10)
```

```python
df = df.dropna(subset=['transactiondate'])
```

```python
df['transactiondate'].head(10)
```

```python
df = df.reset_index(drop=True)
```

```python
df['loginattempts'] = df['loginattempts'].astype(int)
```

```python
df['loginattempts'].dtype
```

```python
# creating fraud flag
df['login_risk'] = df['loginattempts'] > 3
```

```python
df['login_risk'].value_counts()
```

```python
df[df['login_risk'] == True]
```

```python
#“I observed that most users had a single login attempt, but a small portion had multiple attempts (up to 5), 
#which is a strong fraud indicator. I created a login risk flag to identify such suspicious behavior.”
```

```python
# transactionduration
df['transactionduration'].dtype
df['transactionduration'].describe()
```

```python
df['transactionduration'] = df['transactionduration'].astype(int)
df['transactionduration'].dtype
```

```python
#Creating Duration Risk Flag
df['duration_risk'] = df['transactionduration'] > 180
df['duration_risk'].value_counts()
```

```python
df[df['duration_risk'] == True]
```

```python
#I analyzed transaction duration and observed that unusually long durations could indicate suspicious behavior such as system delays or fraudulent attempts. 
#I created a duration-based risk flag to capture such anomalies.
```

```python
df['fraud_flag'] = (
    (df['loginattempts'] > 3) |
    (df['transactionduration'] > 180) |
    (df['transactionamount'] > df['transactionamount'].quantile(0.95))
)
```

```python
df['fraud_flag'].value_counts()
```

```python
df[df['fraud_flag'] == True]
```

```python
##I combined multiple behavioral indicators such as high transaction amount, multiple login attempts, and abnormal transaction duration to 
#create a composite fraud detection flag. This approach is more effective than relying on a single variable.
```
